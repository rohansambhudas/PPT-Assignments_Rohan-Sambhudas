```python
# Question 1
# a
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017")

# Access the "students" collection in the "mydatabase" database
db = client.mydatabase
collection = db.students

# Create a new document
student = {
    "name": "John Doe",
    "age": 20,
    "grade": "A"
}

# Insert the document into the collection
result = collection.insert_one(student)

# Print a success message
if result.acknowledged:
    print("Document inserted successfully.")

# b
from cassandra.cluster import Cluster
from cassandra.query import SimpleStatement

# Connect to Cassandra
cluster = Cluster(['localhost'])
session = cluster.connect()

# Use a keyspace
session.set_keyspace('mykeyspace')

# Define the insert query
query = "INSERT INTO products (id, name, price) VALUES (?, ?, ?)"

# Insert a new record
try:
    session.execute(SimpleStatement(query), (1, 'Product 1', 9.99))
    print("Record inserted successfully.")
except Exception as e:
    print(f"Error inserting record: {str(e)}")

# Close the connection
cluster.shutdown()
```

```python
# Question 2
# a
from pymongo import MongoClient
from datetime import datetime, timedelta

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017")

# Access the "books" collection in the "mydatabase" database
db = client.mydatabase
collection = db.books

# Get the date one year ago from today
one_year_ago = datetime.now() - timedelta(days=365)

# Query the collection for books published in the last year
query = {"published_date": {"$gte": one_year_ago}}
projection = {"title": 1, "author": 1}

cursor = collection.find(query, projection)

# Print the titles and authors of the books
for book in cursor:
    print(f"Title: {book['title']}")
    print(f"Author: {book['author']}")
    print()

# b
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017")

# Access the "customers" collection in the "mydatabase" database
db = client.mydatabase
collection = db.customers

# Validate and insert a new customer document
def insert_customer(customer):
    # Perform validation checks
    if "name" not in customer:
        raise ValueError("Customer name is required.")
    if "email" not in customer:
        raise ValueError("Customer email is required.")

    # Insert the customer document into the collection
    result = collection.insert_one(customer)

    # Print a success message
    if result.acknowledged:
        print("Customer inserted successfully.")

# Example usage
customer_data = {
    "name": "John Doe",
    "email": "johndoe@example.com",
    "phone": "1234567890"
}

try:
    insert_customer(customer_data)
except Exception as e:
    print(f"Error inserting customer: {str(e)}")
```

```python
# Question 3
# a
from pymongo import MongoClient

# Connect to the MongoDB replica set
client = MongoClient("mongodb://node1:27017,node2:27017,node3:27017/?replicaSet=myreplicaset")

# Get the replica set status
status = client.admin.command("replSetGetStatus")

# Print the status of each node
for member in status["members"]:
    if member["stateStr"] == "PRIMARY":
        print(f"Primary Node: {member['name']}")
    elif member["stateStr"] == "SECONDARY":
        print(f"Secondary Node: {member['name']}")

        
# b
from cassandra.cluster import Cluster

# Connect to the Cassandra cluster
cluster = Cluster(['node1', 'node2', 'node3'])

# Retrieve the cluster metadata
metadata = cluster.metadata

# Get the status of each node
for host in metadata.all_hosts():
    print(f"Node: {host.address}, Status: {host.is_up}")
```

```python
# Question 4
# a
from pymongo import MongoClient

# Connect to the MongoDB cluster
client = MongoClient("mongodb://mongos1:27017")

# Enable sharding for the database
client.admin.command("enableSharding", "mydb")

# Shard the collection based on the sharding key
client.admin.command("shardCollection", "mydb.mycol", key={"shardingKeyField": "hashed"})

# Insert documents into the sharded collection
db = client.mydb
col = db.mycol
col.insert_many([
    {"_id": 1, "shardingKeyField": "value1", "otherField": "data1"},
    {"_id": 2, "shardingKeyField": "value2", "otherField": "data2"},
    {"_id": 3, "shardingKeyField": "value3", "otherField": "data3"}
])


# b
from pymongo import MongoClient

# Connect to the MongoDB cluster
client = MongoClient("mongodb://mongos1:27017")

# Access the sharded database and collection
db = client.mydb
col = db.users

# Insert a user document
col.insert_one({"_id": 1, "username": "user1", "name": "John Doe", "age": 25})

# Retrieve a user document by ID
user = col.find_one({"_id": 1})
print(user)
```

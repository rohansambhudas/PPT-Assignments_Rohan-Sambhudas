```python
# Question 5
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='<hive_server_host>', port=<hive_server_port>, username='<username>')

# Create a cursor object to execute HiveQL queries
cursor = conn.cursor()

# Define the HiveQL query to calculate the average salary
query = '''
    SELECT AVG(salary) FROM Employees
'''

# Execute the query
cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Access the average salary from the result
average_salary = result[0]

# Print the average salary
print(f"Average Salary: {average_salary}")

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
# Question 8
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='<hive_server_host>', port=<hive_server_port>, username='<username>')

# Create a cursor object to execute HiveQL queries
cursor = conn.cursor()

# Define the HiveQL query to perform the inner join
query = '''
    SELECT *
    FROM Orders
    INNER JOIN Customers ON Orders.customer_id = Customers.customer_id
'''

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
# Question 9
from pyhive import hive

# Establish a connection to Hive
conn = hive.Connection(host='<hive_server_host>', port=<hive_server_port>, username='<username>')

# Create a cursor object to execute HiveQL queries
cursor = conn.cursor()

# Define the HiveQL query to process JSON data using the SerDe library
query = '''
    SELECT *
    FROM User_Activity_Logs
    ROW FORMAT SERDE 'org.apache.hive.hcatalog.data.JsonSerDe'
    STORED AS TEXTFILE
'''

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Close the cursor and connection
cursor.close()
conn.close()
```
